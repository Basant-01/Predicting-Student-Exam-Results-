# Linear Regression

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.head())
```

```python
print(x_test.head())
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred_train_lr=lr.predict(x_train)
y_pred_test_lr=lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse_train_lr=mean_squared_error(y_train,y_pred_train_lr)
mse_test_lr=mean_squared_error(y_test,y_pred_test_lr)
print('MSE Training :',mse_train_lr)
print('MSE Testing :',mse_test_lr)
```

```python
r2_score_train_lr=r2_score(y_train,y_pred_train_lr)
r2_score_test_lr=r2_score(y_test,y_pred_test_lr)
print('R2 Score Training :',r2_score_train_lr)
print('R2 Score Testing :',r2_score_test_lr)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_train, y_train, color='blue', label='Training Data')
plt.scatter(x_test, y_test, color='green', label='Testing Data')
plt.plot(x_train, y_pred_train_lr, color='red', linewidth=2, label='Linear Regression')
plt.xlabel('Study Time')
plt.ylabel('Final Exam Score')
plt.title('Linear Regression on Student Dataset')
plt.legend()
plt.show()
```

## Logistic Regression


```python
x = data[['AttendanceRate', 'StudyTime']]
y = data['Pass/Fail']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.head())
print(y_train.head())
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(x_train, y_train)
```

```python
y_pred_test_lr = lr_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred_test_lr)
accuracy=accuracy*100
print('Accuracy:', accuracy)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_test_lr))
```

## K-Means clustering

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

features = data[['StudyTime', 'AttendanceRate']]

n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
data['Cluster'] = kmeans.fit_predict(features)
```

```python
# Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(data['StudyTime'], data['AttendanceRate'], c=data['Cluster'], cmap='viridis', marker='o')
plt.xlabel('Study Time')
plt.ylabel('Attendance Rate')
plt.title('Student Clusters based on Study Time and Attendance')
plt.colorbar(label='Cluster')
plt.show()
```

```python
for cluster_id in range(n_clusters):
  cluster_data = data[data['Cluster'] == cluster_id]
  print(f"\nCharacteristics of Cluster {cluster_id}:")
  print(cluster_data[['StudyTime', 'AttendanceRate', 'FinalExamScore', 'Pass/Fail']].describe())
```
